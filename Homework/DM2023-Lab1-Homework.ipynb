### >>> **Exercise 14 (take home):** 
The chart above only contains 300 vocabulary in the documents, and it's already computationally intensive to both compute and visualize. Can you efficiently reduce the number of terms you want to visualize as an exercise. 


```python
# Answer here
# We first clean the data than plot the terms with frequency larger than 20

import re
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px


def data_cleaning(text):
    text = re.sub('\w*\d\w*', '', text) #remove numbers 
    text = re.sub('[@''"".,_-]', '', text) #remove symbols @'".,_-
    return text

X['text'] = X['text'].apply(lambda x: data_cleaning(x))
X_sample = X.sample(n=1000)

cleaned_count_vect = CountVectorizer()
cleaned_counts = cleaned_count_vect.fit_transform(X_sample.text)

print(len(count_vect.get_feature_names_out()))
print(len(cleaned_count_vect.get_feature_names_out()))

cleaned_term_frequencies = np.asarray(cleaned_counts.sum(axis=0))[0]

x_feature_name = []
y_freq = []
for index, number in enumerate(cleaned_term_frequencies):
    if number > 20:
        x_feature_name.append(cleaned_count_vect.get_feature_names_out()[index])
        y_freq.append(number)
        
print("Oringinal size of feature name->",len(count_vect.get_feature_names_out()))

print("Remaining size of feature name->",len(x_feature_name))
#plot

fig = go.Figure(
    data=[
        go.Bar(
            x=x_feature_name,
            y=y_freq,
        )
    ]
)
fig.update_layout(xaxis_tickangle=45)

fig.show()
```
